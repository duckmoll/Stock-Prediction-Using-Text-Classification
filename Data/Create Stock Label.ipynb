```python
import yfinance as yf
import pandas as pd
single_df = yf.download("SPY", 
                      start='2014-01-01', 
                      end='2021-11-11', 
                      progress=False)
single_df = single_df[["Adj Close"]]
single_df.columns = ["SPY"]
single_df.index = pd.to_datetime(single_df.index)
single_df
```

```python
data = pd.read_csv("sp_prices.csv", index_col=0)
data.index = pd.to_datetime(data.index)
data.head(3)
```

```python
single_df = single_df.merge(data, left_index=True, right_index=True, how="right")
single_df = single_df[["SPY"]]
single_df
```

```python
returns = data.pct_change()
returns = returns.shift(-5).drop(returns.index[-1])
returns
```

```python
single_df_return = single_df.pct_change()
single_df_return = single_df_return.shift(-1).dropna()

pre_label = pd.DataFrame()
for col in returns.columns:
    pre_label[col] = returns[col] - single_df_return["SPY"]
pre_label[pre_label > 0] = 1
pre_label[pre_label <= 0] = 0
```

```python
labels_1d = pre_label.stack().reset_index()
labels_1d.columns = ["date", "ticker", "label"]
labels_1d.to_csv("labels_5d.csv")
```
