```python
from secedgar.core import CompanyFilings, FilingType, DailyFilings
from secedgar.exceptions import NoFilingsError, EDGARQueryError
from datetime import date
import datetime
import time
import requests
from aiohttp import ClientResponseError

urls_dict = {}
start = date(2015, 1, 9)
def daily_filter(filing_entry):
    return filing_entry.cik in ciks and filing_entry.form_type.lower() != "10-k" and filing_entry.form_type.lower() != "10-q"
while start <= date(2021, 10, 29):
    print(start)
    my_filings = DailyFilings(date=start,
                               user_agent="joey",
                              entry_filter=daily_filter
                             )
    try:
        urls_dict[start] = my_filings.get_urls_safely()
    except (NoFilingsError, EDGARQueryError):
        start = start + datetime.timedelta(days=1)
        continue
    start = start + datetime.timedelta(days=1)
```

```python
download = urls_dict
for key in list(download.keys()):
    n_key = str(key)
    download[n_key] = download[key]
    del download[key]
```
