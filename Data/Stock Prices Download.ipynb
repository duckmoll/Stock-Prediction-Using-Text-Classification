```python
import pandas as pd
import requests
import json
```

```python
data = pd.read_csv("sec_file_mapping.csv", index_col=0)
data["ticker"] = data["ticker"].str.replace("-", ".")
data
```

```python
import yfinance as yf
import pandas as pd
all_df = pd.DataFrame()
tickers = list(set(data["ticker"]))
for index, ticker in enumerate(tickers):
    print(index)
    single_df = yf.download(ticker, 
                          start='2014-01-01', 
                          end='2021-11-11', 
                          progress=False)
    single_df = single_df[["Adj Close"]]
    single_df.columns = [ticker]
    all_df = all_df.merge(single_df, left_index=True, right_index=True, how="outer")
all_df
```

```python
all_df.isnull().sum()
```

```python
all_df_cleaned = all_df.dropna(axis=1,thresh=51)
all_df_cleaned = all_df_cleaned.fillna(method="ffill")
all_df_cleaned
```

```python
all_df_cleaned.to_csv("sp_prices.csv")
```
