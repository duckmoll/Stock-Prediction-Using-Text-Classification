```python
import pandas as pd
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
```

```python
data = pd.read_csv("/content/drive/MyDrive/nlp/rnn.csv")
for form in ["4", "424B2", "8-K", "SC 13G/A", "FWP", "3",
             "SC 13G", "425", "DEFA14A", "4/A", "DEF 14A", 
             "UPLOAD", "11-K", "5", "CORRESP", "424B5"]:
  form_data = data[data["form_type"] == form]
  print(form)
  print(confusion_matrix(form_data["label"], form_data["preds"]))
  print(classification_report(form_data["label"], form_data["preds"]))
```

```python
print(confusion_matrix(data["label"], data["preds"]))
print(classification_report(data["label"], data["preds"]))
```
