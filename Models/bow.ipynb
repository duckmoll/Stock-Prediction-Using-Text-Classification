```python
import pandas as pd
```

```python
X_train = pd.read_csv("/content/drive/MyDrive/nlp/training_content_all.csv").dropna()
X_train = X_train[X_train["form_type"] != "4"]
print(len(X_train))
y_train = X_train["label"]
X_train.head(3)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
import gc
vectorizer=CountVectorizer(min_df=100, max_features=8000)
wd_train = vectorizer.fit_transform(X_train["content"])
del X_train
gc.collect()
wd_train = pd.DataFrame(wd_train.toarray(),
                        columns=vectorizer.get_feature_names())
wd_train.head(3)
```

```python
from sklearn.naive_bayes import MultinomialNB

classifier = MultinomialNB()
classifier.fit(wd_train, y_train)
```

```python
del wd_train, y_train
gc.collect()
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report

X_test = pd.read_csv("/content/drive/MyDrive/nlp/test_content_all.csv").dropna()
X_test = X_test[X_test["form_type"] != "4"]
print(len(X_test))
wd_test = vectorizer.transform(X_test["content"])
pred = classifier.predict(wd_test.toarray())
X_test["preds"] = pred
y_test = X_test["label"]
# bn_matrix = pd.DataFrame(confusion_matrix(y_test, pred),
#                          index=classifier.classes_,
#                          columns=classifier.classes_)
# print(bn_matrix)
# print(classification_report(y_test, pred))
```
